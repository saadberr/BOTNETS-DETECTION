## proto :

```python
proto = train['proto'].value_counts()
proto
```

```python
# There is 131 different protocol, so we will visualize just top 10 frequent protocols.
```

```python
proto = proto[:10]
plt.figure(figsize=(8,6))
plt.bar(proto.index, proto.values)
plt.xlabel('Protocol')
plt.ylabel('Freq')
plt.show()
```

## service : 

```python
srv = train['service'].value_counts()
srv
```

```python
plt.figure(figsize=(8,6))
plt.bar(srv.index, srv.values)
plt.xlabel('Service')
plt.ylabel('Freq')
plt.show()
```

## Attack categories :

```python
atc = train['attack_cat'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(atc.index, atc.values)
plt.xticks(rotation=90)
plt.xlabel('Attack category')
plt.ylabel('Freq')
plt.show()
```

So there is 10 categories, and our data isn't balanced, so we will visualize wether an activity is normal or not

```python
sns.countplot(train['label'])
plt.xlabel('Attack category')
plt.ylabel('Freq')
plt.show()
```

So data is balanced between 0 and 1 (normal or not). 

## Correlations :

```python
plt.figure(figsize=(15,15))
sns.heatmap(train.corr())
plt.show()
```

Many features wich are correlated with each other.
